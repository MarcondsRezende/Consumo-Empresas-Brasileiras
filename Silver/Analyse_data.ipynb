```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/marco/OneDrive/Documentos/GitHub/Consumo-Empresas-Brasileiras/Silver/dados_jf.csv")
df
```

```python
df['Todas Vazias'] = (df['DDD 1'].isnull() & df['TELEFONE 1'].isnull() & df['DDD 2'].isnull() & df['TELEFONE 2'].isnull()).astype(int)
len(df)
```

```python
df = df[df['Todas Vazias'] != 1]
len(df)
```

```python
df.info()
```

```python
df = df.drop(['result', 'Todas Vazias'], axis=1)
df
```

```python
# def para mapear

def mapear_porte(valor):
    if valor == 3:
        return 'Empresa de Pequeno Porte'
    elif valor == 1:
        return 'Micro Empresa'
    elif valor == 5:
        return 'Demais Empresas'
    else:
        return 'Não Informado'
```

```python
df['Porte Empresa'] = df['PORTE DA EMPRESA'].apply(mapear_porte)
df.drop('PORTE DA EMPRESA', axis=1)
```

```python
df.drop_duplicates(inplace=True)
len(df)
```

```python
df
```

```python
df = df.sort_values(by='CNPJ BÁSICO')
df
```

```python
df['TELEFONE 1'] = df['TELEFONE 1'].astype(str)
df['TELEFONE 2'] = df['TELEFONE 2'].astype(str)
```

```python
df['Pri.Num 1'] = df['TELEFONE 1'].str.slice(0,1)
df['Pri.Num 2'] = df['TELEFONE 2'].str.slice(0,1)
```
